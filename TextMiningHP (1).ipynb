Primeramente, realizaremos unos primeros análisis. Contaremos el numero de caracteres que existen en el texto con el que vamos a trabajar.

```python
print('El número total de caracteres es de : {}'.format(len(texto)))
```

### Analizamos cual es el idioma del texto ###

Para ello usaremos la libreria textacy.

```python
print("Idioma detectado por Textacy: ",textacy.text_utils.detect_language(texto))
```

Nos dice que el idioma del texto detectado es el ingles, por lo que acierta de lleno.

### Separamos el texto en frases

A continuación separaremos el texto en frases como se nos indica. Para ello usaremos la libreria nltk

En este caso especificamos que queremos la frase numero 4 del fragmento que hemos elegido.

```python
frases = nltk.sent_tokenize(texto)
print(frases[4])
```

### División del texto por tokens

En este apartado, dividiremos el texto por tokens. Los tokens se refieren a palabras.

Especificamos que nos coja desde la sexta palabra hasta la decimo octava

```python
tokens = nltk.word_tokenize(texto)
tokens[6:18]
```

### Obtencion de lemas de los tokens anteriores

Aqui lo que haremos será recorrer palabra por palabra descomponiendo la forma normal de cada una de ellas.

lemmatizer = WordNetLemmatizer()

```python
for palabra in tokens:
    print(lemmatizer.lemmatize(palabra))
```
